# One sided distribution: importance calculation

Domains A and B contain digits with class (0, 1). They have have the following distributions:

```python
p_A0 = 0.4
p_A1 = 0.6

p_B0 = 0.5
p_B1 = 0.5
```

The multiplication factor for each of the classes for domain A is as following:

```python
m0 = p_B0 / p_A0
m1 = p_B1 / p_A1

m0, m1
```

Normalize so that we can weight or sample, these are now the importances:

```python
t = m0+m1

nm0 = m0 / t
nm1 = m1 / t

nm0, nm1
```

Plot the importances based on the distributions:

```python
import matplotlib.pyplot as plt

import numpy as np
```

```python
p_A0s = np.linspace(0, 1, 101)[1:-1]
p_A1s = 1 - p_A0s

p_B0s = 0.5
p_B1s = 0.5
```

```python
m0s = p_B0s / p_A0s
m1s = p_B1s / p_A1s

ts = m0s + m1s

nm0s = m0s / ts
nm1s = m1s / ts
```

```python
plt.plot(nm0s, p_A0s)
plt.plot(nm1s, p_A0s)
plt.title('normalized importance per class over source distribution')
plt.xlabel('p_A0')
plt.ylabel('normalized importance for the class')
plt.legend(['0', '1'])
plt.show()
```
